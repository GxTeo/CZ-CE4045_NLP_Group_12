```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from wordcloud import WordCloud
```

```python
#Read json file via pandas
df = pd.read_json('dataset.json')
df.info()
```

```python
df.head(10)
```

Pre-Processing

```python
#Convert to text to lower-case
df["reviewDescription_lower_case"] = df["reviewDescription"].str.lower()
df.head(10)
```

```python
count_vectorizer = CountVectorizer(ngram_range=(1, 2), stop_words = 'english',
                                    token_pattern="\\b[a-z][a-z]+\\b", lowercase=True,
                                    max_df = 0.6, max_features= 4000)

tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), stop_words='english',
                                   token_pattern= "\\b[a-z][a-z]+\\b", lowercase=True,
                                    max_df = 0.6, max_features=4000)

cv_data = count_vectorizer.fit_transform(df["reviewDescription_lower_case"])
tfidf_data = tfidf_vectorizer.fit_transform(df["reviewDescription_lower_case"])
```

```python
for_wordcloud = count_vectorizer.get_feature_names()
for_wordcloud = for_wordcloud
for_wordcloud_str = ' '.join(for_wordcloud)

wordcloud = WordCloud(width=800, height=400, background_color='black', min_font_size = 7).generate(for_wordcloud_str)

plt.figure(figsize=(20, 10), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
